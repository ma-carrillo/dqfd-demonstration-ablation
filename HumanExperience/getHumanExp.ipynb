## Flappy Bird

```python
import pickle
import time

import gymnasium as gym
import flappy_bird_gymnasium
import pygame


NUM_EPISODES = 50
OUTPUT_FILE = "flappybird_human_demos.pkl"


def main():
    env = gym.make("FlappyBird-v0", render_mode="human")

    pygame.init()
    clock = pygame.time.Clock()

    all_episodes = []

    print("Controls:")
    print("  SPACE = flap")
    print("  ESC   = quit current run")
    print()

    try:
        for episode_idx in range(NUM_EPISODES):
            obs, info = env.reset()
            done = False
            episode_data = []
            step_count = 0
            episode_reward = 0.0

            print(f"Starting episode {episode_idx + 1}/{NUM_EPISODES}...")
            time.sleep(1.0)

            while not done:
                action = 0

                # Read keyboard events
                for event in pygame.event.get():
                    if event.type == pygame.QUIT:
                        done = True
                        break
                    elif event.type == pygame.KEYDOWN:
                        if event.key == pygame.K_SPACE:
                            action = 1
                        elif event.key == pygame.K_ESCAPE:
                            done = True
                            break

                if done:
                    break

                next_obs, reward, terminated, truncated, info = env.step(action)
                done = terminated or truncated

                episode_data.append(
                    {
                        "state": obs,
                        "action": action,
                        "reward": reward,
                        "next_state": next_obs,
                        "done": done,
                    }
                )

                obs = next_obs
                episode_reward += reward
                step_count += 1

                # Limit loop speed a bit so keyboard input is manageable
                clock.tick(30)

            if episode_data:
                all_episodes.append(episode_data)
                print(
                    f"Episode {episode_idx + 1} saved "
                    f"({step_count} steps, reward={episode_reward:.2f})"
                )
            else:
                print(f"Episode {episode_idx + 1} discarded (no data recorded).")

    finally:
        env.close()
        pygame.quit()

    with open(OUTPUT_FILE, "wb") as f:
        pickle.dump(all_episodes, f)

    print()
    print(f"Saved {len(all_episodes)} episodes to: {OUTPUT_FILE}")


if __name__ == "__main__":
    main()
```

## Snake

```python
import numpy as np
np.bool8 = bool  # only if needed for your Gym/Numpy setup

import gym
import gym_snake_game
import pygame
import pickle

NUM_EPISODES = 50
OUTPUT_FILE = "snake_human_demos.pkl"


def get_action_from_key(key):
    if key == pygame.K_UP:
        return 0
    if key == pygame.K_RIGHT:
        return 3
    if key == pygame.K_DOWN:
        return 1
    if key == pygame.K_LEFT:
        return 2
    return None


def main():
    env = gym.make("Snake-v0", render_mode="human", width=15, height=15)

    pygame.init()
    clock = pygame.time.Clock()

    dataset = []

    for ep in range(NUM_EPISODES):
        state, info = env.reset()
        done = False
        episode_data = []

        # keep moving in the last chosen direction
        action = 1  # start moving right, change if needed

        print(f"Episode {ep + 1}")

        while not done:
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    done = True
                elif event.type == pygame.KEYDOWN:
                    new_action = get_action_from_key(event.key)
                    if new_action is not None:
                        action = new_action

            if done:
                break

            next_state, reward, terminated, truncated, info = env.step(action)
            done = terminated or truncated

            episode_data.append({
                "state": state,
                "action": action,
                "reward": reward,
                "next_state": next_state,
                "done": done,
            })

            state = next_state
            clock.tick(10)

        if episode_data:
            dataset.append(episode_data)
            print("Episode length:", len(episode_data))

    env.close()
    pygame.quit()

    with open(OUTPUT_FILE, "wb") as f:
        pickle.dump(dataset, f)

    print("Saved dataset to:", OUTPUT_FILE)


if __name__ == "__main__":
    main()
```
